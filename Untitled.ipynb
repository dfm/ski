```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as pl
```

```python
rng = (0, 2)
x = np.sort(np.random.uniform(rng[0], rng[1], 100))
x0 = np.linspace(rng[0], rng[1], 40000)
y = np.sin(x0)
```

```python
pl.plot(x0, y, ".k")
```

```python
hi = np.searchsorted(x0, x)
lo = hi - 1

lo = np.clip(lo, 0, len(x0) - 1)
hi = np.clip(hi, 0, len(x0) - 1)

m = hi > lo
w = np.zeros(len(m))
w[m] = (x[m] - x0[lo][m]) / (x0[hi][m] - x0[lo][m])
```

```python
W = np.zeros((len(x), len(x0)))

W[np.arange(len(x)), hi] += w
W[np.arange(len(x)), lo] += 1. - w

# m = w <= 0.
# W[np.arange(len(x))[m], lo[m]] = 1.0
# m = w >= 1.
# W[np.arange(len(x))[m], hi[m]] = 1.0
```

```python
W[0]
```

```python
Kuu = np.exp(-0.5 * (x0[:, None] - x0[None, :])**2)
Kski = np.dot(W, np.dot(Kuu, W.T))
Kxx = np.exp(-0.5 * (x[:, None] - x[None, :])**2)
```

```python
pl.imshow(Kski, cmap="gray", interpolation="nearest")
pl.colorbar();
```

```python
pl.imshow(np.abs(Kxx - Kski), cmap="jet", interpolation="nearest")
pl.colorbar();
```

```python
pl.plot(x0, y, "ok")
pl.plot(x, np.dot(W, y), ".g")
```
